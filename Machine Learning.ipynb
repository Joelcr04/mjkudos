```python
def plot_iris(iris,col1,col2):
    
    import seaborn as sns
    import matplotlib.pyplot as plt
    
    sns.lmplot(x=col1, y=col2,
              data=iris,
              hue="Species",
              fit_reg=False)
    plt.xlabel(col1)
    plt.ylabel(col2)
    
    plt.title("Iris Species")
    plt.show()
```

```python
plot_iris(iris.data,"Petal_width","Sepal_length")
plot_iris(iris.data,"Sepal_width","Sepal_length")
```

# Splitting the data into Training and testing set

```python
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
np.random.seed(123)
```

```python
m=np.asmatrix(iris_scaled)
m
```

```python
iris_split=train_test_split(m,test_size=75)
```

```python
iris_split
```

```python
np.ravel(iris_split[1][:4])
```

```python
X_train=iris_split[0][::4,]
Y_train=np.ravel(iris_split[0][:4])
```

```python
X_test=iris_split[1][::4,]
Y_test=np.ravel(iris_split[1][::4,])
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN_Mod=KNeighborsClassifier(n_neighbors=15)
```

```python
KNN_Mod.fit(X_train,Y_train)
```

```python
iris_test=pd.DataFrame(X_test,columns=colnames)
iris_test
```

```python
iris_test["predicted"]=KNN_Mod.predict(X_test)
iris_test
```

```python
iris_test["correct"]=[1 if x==z else 0 for \
                    x,z in zip(iris_test["predicted"],Y_test)]
```
